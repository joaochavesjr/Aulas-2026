```python
import pandas as pd
```

```python
df = pd.read_csv("exemplo.csv")
```

```python
df
```

```python
selected_columns = df[['Date', 'Product', 'Sales']]
print(selected_columns.head())
```

```python
accessories_sales = df[df['Category'] == 'Accessories']
print(accessories_sales)
```

```python
high_sales_electronics = df[(df['Sales'] > 1500) & (df['Category'] ==
'Electronics')]
print(high_sales_electronics)
```

```python
# Adicionar uma coluna com a receita total (Sales * Quantity)
df['Total_Revenue'] = df['Sales'] * df['Quantity']
print(df.head())
```

```python
# Transformar a coluna 'Sales' para representar valores em reais (fator deconversão fictício)
conversion_factor = 5.5
df['Sales_BRL'] = df['Sales'] * conversion_factor
print(df[['Date', 'Sales', 'Category', 'Quantity']])
```

```python
# Renomear a coluna 'Quantity' para 'Units_Sold'
#df.rename(columns={'Quantity': 'QTDE'}, inplace=True)
df.rename(columns={'QTDE': 'Quantity'}, inplace=True)
print(df.head())
```

```python
# Agrupar por 'Category' e calcular a soma das venda
categoory_sales = df.groupby('Category')['Sales'].sum()
categoory_sales
```

```python
# Agrupar por 'Date' e calcular a média das vendas
daily_sales_mean = df.groupby('Date')['Sales'].mean()
print(daily_sales_mean)
```

```python
# Agrupar por 'Product' e calcular soma e média das vendas
product_sales = df.groupby('Product').agg({'Sales': ['sum', 'mean']})
print(product_sales)
```
