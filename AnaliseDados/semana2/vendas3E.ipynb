```python
import pandas as pd
```

```python
df = pd.read_excel('vendas.xlsx')
```

```python
# Exemplo python funções string puro
texto = '  tEXto BLA  '
type(texto)
texto.replace('BLA','')
```

```python
df['Resposta'] = df['Resposta'].str.lower().str.strip().str.title()
```

```python
df['Descricao'] = df['Descricao'].str.replace('#$%!', '')
```

```python
df['Preco_Compra']
```

```python
df['Preco_Compra'] = df['Preco_Compra'].str.replace('R$ ','')
df['Preco_Compra'] = df['Preco_Compra'].str.replace('.', '').str.replace(',', '.')
df['Preco_Compra'] = pd.to_numeric(df['Preco_Compra'])
```

```python
df['Quantidade'] = df['Quantidade'].astype(str).str.replace('items', '')
```

```python
df['Preco_Compra'] = df['Preco_Compra'].fillna(df['Preco_Unitario']*df['Quantidade'].astype(int))
```

```python
# Filtro para apresnetar os valores do campo Nome nulos
df[df['Nome'].isnull()]
```

```python
df['Nome'] = df['Nome'].fillna('Desconhecido')
```

```python
df['Data_Nascimento'] = pd.to_datetime(df['Data_Nascimento'], format='%d/%m/%Y', errors='coerce')
```
