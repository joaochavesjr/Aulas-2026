### Importar bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

### Gráfico 3D

```python
# Geração de dados aleatórios em 3D
np.random.seed(42)
X = np.dot(np.random.rand(3, 3), np.random.rand(3, 200)).T

# Visualização dos dados 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c='blue', marker='o')
ax.set_title('Dados Originais em 3D')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
plt.show()
```

### Gráfico 2D após aplicação de redução de dimensionalidade (PCA)

```python
# Aplicação do PCA para reduzir de 3D para 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualização dos dados 2D após a redução de dimensionalidade
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='red', marker='o')
plt.title('Dados em 2D após PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()
```
